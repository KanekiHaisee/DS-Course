MAP FUNCTION :

```python
l=[2,3,4,5,6]
```

```python
def test(l):
    l1=[]
    for i in l:
        l1.append(i**2)
    return l1
```

```python
test(l)
```

```python
def sq(x):
    return x**2
```

```python
list(map(sq,l))
```

```python
list(map(lambda x : x**2,l))
```

```python
l1=[2,1,3,4,5]
l2=[6,7,8,9,0]

list(map(lambda x,y : x+y,l1,l2)) #l1 and l2 are iterables
#shows result in list(maps the function (lambda function takes input x and y as 2 numbers from lists :
#returns their addition , list1 from where x is taken and l2 from where y is taken))
```

```python
s="deva"
list(map(lambda s : s.upper() ,s))
```

REDUCE : not available by default

```python
from functools import reduce
```

```python
l=[1,2,3,4,5]
```

```python
reduce(lambda c,y:c+y,l) #only 2 parameters by default
```

```python
reduce(lambda c,y:c+y,[1]) #Exception
```

```python
reduce(lambda x,y : x*y,l)
```

```python
l
```

```python
reduce(lambda x,y : x if x>y else y , l)
```
